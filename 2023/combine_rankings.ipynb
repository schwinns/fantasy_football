```python
import pandas as pd
```

```python
preds = pd.read_csv('/content/drive/MyDrive/fantasy_football/predictions_2023.csv')
preds = preds[['Player', 'ENetRk', 'SVRRk', 'KNNRk', 'GBRRk', 'DNNRk']]
preds
```

```python
all_ranks = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vSwnO49IGl-uGORj7NZYYnuKMWEQK6RcKi1gcTqmGnu8i7gb_SYPvPy-gAyu6h7tt10SXDTqCWbm0eB/pub?gid=0&single=true&output=csv')
nbc = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQWvVBy6pO8Y7zbUspY_XWW2a81lj2tWcqEU3FCo-Qb_Mk4K1ULzhLnXLRvEFCKnG1oPTYnMX7GsiFf/pub?gid=1324057779&single=true&output=csv')
the192 = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vRVyguyKcZ4wEBXhSjo7Oc_vKro7VhMZV9uc9wf1ainNshli0Kye0JZSoSWLS7wZ99LZw02H2RExyEJ/pub?gid=0&single=true&output=csv')
```

```python
tmp = all_ranks.merge(preds, how='left', left_on='Player', right_on='Player')
tmp = tmp.merge(nbc, how='left', left_on='Player', right_on='Player')
tmp.rename(columns={'Rank' : 'NBC Berry'}, inplace=True)
all_ranks = tmp.merge(the192, how='left', left_on='Player', right_on='Player')
all_ranks
```

```python
all_ranks.to_csv('/content/drive/MyDrive/fantasy_football/all_ranks.csv', index=False)
```
